```python
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
df_input = df[['cgpa','iq']]
```

```python
df_input.head()
```

```python
df_output = df[['placement']]
```

```python
df_output.head()
```

```python
from sklearn.model_selection import train_test_split

train_test_split(df_input,df_output,test_size=0.1)
```

```python
ip_train,ip_test,op_train,op_test = train_test_split(df_input,df_output,test_size=0.1)
```

```python
op_train
```

```python
ip_train
```

```python
ip_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
ip_train = scaler.fit_transform(ip_train)
```

```python
ip_train
```

```python
ip_test = scaler.transform(ip_test)
```

```python
ip_test
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
```

```python
clf.fit(ip_train, op_train)
```

```python
op_pred = clf.predict(ip_test)
```

```python
op_pred
```

```python
op_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(op_pred,op_test)
```

```python
accuracy_rounded = round(accuracy, 2)
accuracy_rounded
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(ip_train, op_train.values.ravel(), clf=clf, legend=2)
```
